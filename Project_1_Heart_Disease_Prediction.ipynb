Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
#loading the dataset
heart_data = pd.read_csv('/content/heart_disease_data.csv')
```

```python
#printing first 5 rows of dataset
heart_data.head()
```

```python
#last 5 rows of the dataset
heart_data.tail()
```

```python
#number of rows and coloumns in the dataset
heart_data.shape
```

```python
#Now getting some infor regarding the dataset
heart_data.info()
```

```python
#Now check for missing values
heart_data.isnull().sum()
```

```python
#stattical measures of the heart
heart_data.describe()
```

```python
#checking the distribution of Target Variables
heart_data['target'].value_counts()
```

1 -> Defective Heart

0 -> Healthy Heart


Splitting the dataset(Features and Target)

```python
X = heart_data.drop(columns='target', axis = 1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training and Test Set

```python
X_train,  X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,stratify =Y, random_state =2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Tranining


Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Set
model.fit(X_train, Y_train)
```

Model Evalulation

Accuracy Score

```python
#accuracy on training set
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training Data : ', training_data_accuracy)
```

```python
#accuracy on test set
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data : ', test_data_accuracy)
```

Buliding a predctive System

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)
#change the input data to numpy array
input_data_as_numpy_array = np.asarray(input_data)
#reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
  print('The person does not have Heart Disease')
else:
  print('The person has Heart Disease')
```
